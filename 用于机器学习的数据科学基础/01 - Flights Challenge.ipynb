## Solution

### 清理数据 Clean Data

#### 检查空值 Check Null Values

```python
df_flights.isnull().sum()
```

从运行结果可知，在`DepDelay`中有2761个空值，接下来对空值进行处理。

```python
df_flights.DepDel15 = (df_flights.DepDelay >= 15) + 0 # 将布尔值转换为数字
df_flights.isnull().sum()
```

#### 检查异常值 Check Outliers

```python
from matplotlib import pyplot as plt

def ShowDistribution(data, dataname):
    min_val = data.min()
    max_val = data.max()
    mean_val = data.mean()
    med_val = data.median()
    mod_val = data.mode()[0]

    print('{:s} Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(dataname,min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

    # Plot the histogram   
    ax[0].hist(data)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(data, vert=False)
    ax[1].set_xlabel('Value')

    # Add a title to the Figure
    fig.suptitle('{:s} Distribution'.format(dataname))

    # Show the figure
    plt.show()

ShowDistribution(df_flights.DepDelay, 'DepDelay')
ShowDistribution(df_flights.ArrDelay, 'ArrDelay')
```

可见在`DepDelay`和`ArrDelay`中都有有大量异常值，接下来对异常值进行处理。

##### 处理DepDelay异常值

```python
dq01 = df_flights.DepDelay.quantile(0.01)
dq95 = df_flights.DepDelay.quantile(0.95)
print("q01:{0}, q95:{1}".format(dq01, dq95))
normal_depdelay = df_flights[df_flights.DepDelay > dq01].DepDelay
normal_depdelay = normal_depdelay[normal_depdelay < dq95]
ShowDistribution(normal_depdelay, 'DepDelay')
# 尽管已经相对理想，但仍有大量异常值直接排除21以上值
normal_depdelay = normal_depdelay[normal_depdelay < 21]
ShowDistribution(normal_depdelay, 'DepDelay')
```

将处理过DepDelay的数据保存至新变量`df_flights_ddclean`中。

```python
df_flights_ddclean = df_flights[df_flights.DepDelay < 21]
df_flights_ddclean = df_flights_ddclean[df_flights_ddclean.DepDelay > -12]
```

##### 处理ArrDelay异常值

```python
aq01 = df_flights.ArrDelay.quantile(0.01)
aq95 = df_flights.ArrDelay.quantile(0.95)
print("q01:{0}, q95:{1}".format(aq01, aq95))
normal_arrdelay = df_flights[df_flights.ArrDelay > aq01].ArrDelay
normal_arrdelay = normal_arrdelay[normal_arrdelay < aq95]
ShowDistribution(normal_arrdelay, 'ArrDelay')
# 仍有较多异常值，排除36以上值
normal_arrdelay = normal_arrdelay[normal_arrdelay <= 30]
ShowDistribution(normal_arrdelay, 'ArrDelay')
```

##### 综合处理

```python
df_flights_clean = df_flights_ddclean[df_flights_ddclean.ArrDelay <= 30]
df_flights_clean = df_flights_clean[df_flights_clean.ArrDelay >= -33]

ShowDistribution(df_flights_clean.DepDelay, 'DepDelay')
ShowDistribution(df_flights_clean.ArrDelay, 'ArrDelay')
```

由输出可见处理后数据集基本无异常值。

### 探索数据 

#### 汇总统计信息

```python
df_flights_clean.describe()
```

#### **DepDelay**和**ArrDelay**列的分布

```python
ShowDistribution(df_flights_clean.DepDelay, 'DepDelay')
ShowDistribution(df_flights_clean.ArrDelay, 'ArrDelay')
```

#### 数据分析

##### 平均起飞和到达延误

```python
print('平均起飞延误为{:.2f}分钟'.format(df_flights_clean.DepDelay.mean()))
print('平均到达延误为{:.2f}分钟'.format(df_flights_clean.ArrDelay.mean()))
```

##### 承运人的延误表现

```python
df_flights_clean.boxplot('DepDelay',by='Carrier')
df_flights_clean.boxplot('ArrDelay',by='Carrier')
```

##### 不同星期几的到达延误

```python
df_flights_clean.boxplot('DepDelay',by='DayOfWeek')
```

##### 起飞机场对起飞延误的影响

```python
df_flights_clean.boxplot('ArrDelay',by='OriginAirportName', figsize=(20,10), rot=90)
```

##### 晚点与准时起飞对到达延误的影响

```python
df_flights_clean.boxplot('ArrDelay',by='DepDel15')
```

易得准点起飞的平均到达延误时间更短，且准点与晚点起飞的到达延误相差较大。

##### 哪条航线有最多的晚点到达

```python
routes  = pd.Series(df_flights_clean['OriginAirportName'] + ' --> ' + df_flights_clean['DestAirportName'])
df_flights_clean = pd.concat([df_flights_clean, routes.rename("Route")], axis=1)

# Group by routes
route_group = df_flights_clean.groupby(df_flights_clean.Route)
pd.DataFrame(route_group['ArrDel15'].sum()).sort_values('ArrDel15', ascending=False)
```

##### 哪条航线的平均到达延误最高

```python
pd.DataFrame(route_group['ArrDelay'].mean()).sort_values('ArrDelay', ascending=False)
```
